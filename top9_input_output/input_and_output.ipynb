### Writing data to a .csv file

### Reading data in Excel sheet

```python
# Reading single sheet excel sheet is as similar as reading csv
pd.read_excel("Data - Single Worksheet.xlsx")
```

```python
# By default it reads the first sheet
pd.read_excel("Data - Multiple Worksheets.xlsx")
```

```python
# To read specific sheet
# We can access sheets by index or sheets names
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=1)
```

```python
# To read mutiple sheets
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=[0, "Data 2"])
```

```python
# To read all sheets without specifically mention
pd.read_excel("Data - Multiple Worksheets.xlsx", sheet_name=None)
```

### Writing to an Excel

```python
url = "https://data.cityofnewyork.us/api/views/u35m-9t32/rows.csv"
data = pd.read_csv(url)
data
```

```python
first_1000 = data.iloc[0:1000]
first_1000
```

```python
next_1000 = data.iloc[1001:]
next_1000
```

```python
# Creating excel writer object
excel_file = pd.ExcelWriter("community.xlsx")
```

```python
first_1000.to_excel(excel_file, sheet_name="first_1000", index=False)
```

```python
next_1000.to_excel(excel_file, sheet_name="rest_of_the_data", index=False, columns=["City", "State", "Postcode"])
```

```python
excel_file.save()
```
